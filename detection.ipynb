```python
def draw_facebox(filepath, result_list):
  # loading the image
  data = plt.imread(filepath)
  # plotting the image using matplotlib

  plt.imshow(data) # gets the original image for face detection
  ax = plt.gca() # plot each box
  for result in result_list:
    # get coordinates
    x, y, width, height = result['box']
    # create the shape
    rect = plt.Rectangle((x, y), width, height, fill=False, color='blue')
    # draw the box
    ax.add_patch(rect)
    # show the plot
  plt.show()
```

```python
import matplotlib.pyplot as plt
from mtcnn import MTCNN
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle


# load images
filepath = "/content/drive/MyDrive/mayank-khare.jpeg" #single face

#reading image in matplotlib
pixels = plt.imread(filepath)

#loading the detector function using MCNN
detector = MTCNN()

faces = detector.detect_faces(pixels) # detecting faces in the image

draw_facebox(filepath, faces) # display faces on the original image
```

```python
#multi face detection example
filepath_multi = "/content/drive/MyDrive/modi_bollyselfie001.jpg" #mutliple faces

pixels_multi = plt.imread(filepath_multi)

faces_multi = detector.detect_faces(pixels_multi)

faces_multi = detector.detect_faces(pixels_multi) # detecting faces in the image

draw_facebox(filepath_multi, faces_multi) # display faces on the original image

```

```python
#cropping the detected faces for the model
def crop_faces(filename, result_list):
    data = plt.imread(filename)
    cropped_faces = []
    for result in result_list:
        x, y, width, height = result['box']
        cropped_face = data[y:y+height, x:x+width]
        cropped_faces.append(cropped_face)
    return cropped_faces
```

```python
cropped_faces = crop_faces(filepath_multi, faces_multi)
for i, face in enumerate(cropped_faces):
    plt.subplot(1, len(cropped_faces), i+1)
    plt.imshow(face)
    plt.axis('off')
plt.show()
```
